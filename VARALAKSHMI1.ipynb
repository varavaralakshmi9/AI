```python
import pandas as pd
#read csv file into Dataframe
df=pd.read_excel('/content/drive/MyDrive/Data Join vs Data Blending.xlsx')
df['SALARY'] = df['SALARY'].map(lambda x:x*3.10)
print(df)
```

```python
filtered_df = df[df['EMPID'] > 20]
print(filtered_df)
```

```python
from functools import reduce
if not filtered_df['EMPID'].empty:
   total = reduce(lambda x,y:x+y,filtered_df['EMPID'])
   print(total)
else:
  print("The filtered DataFrame is empty,cannot calculate total EMPID.")
```
