```python
import pandas as pd
import scipy.stats as stats
df=pd.read_csv('/content/drive/MyDrive/drive/awards.csv')
award_id_df = df['award_id'].values
t_stat, p_val = stats.ttest_1samp(award_id_df, popmean=0)
print(f" t-test: t_stat={t_stat},p_val={p_val}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
df=pd.read_csv('/content/drive/MyDrive/drive/awards.csv')
contingency_tabel = pd.crosstab(df['award_id'],df['award_date'])
chi2, p, dof,ex = chi2_contingency(contingency_tabel)
print(f"chi-Square Test of Independence: chi2={chi2}, p={p}, dof={dof}, expected={ex}")
```

```python
import pandas as pd
from scipy import stats
df=pd.read_csv('/content/drive/MyDrive/drive/awards.csv')
u_statistic, p_value = stats.mannwhitneyu(df['award_id'],df['award_date'])
print(f"wilcoxon rank-sum test: u_statistic={u_statistic}, p_value={p_value}")
```

```python
import pandas as pd
from scipy import stats
df = pd.read_csv('/content/drive/MyDrive/drive/awards.csv')
groups = df.groupby('award_date')['award_id'].apply(list)
f_value, p_value = stats.f_oneway(*groups)
print(f"F-value: {f_value}")
print(f"p-value: {p_value}")
```

```python
import pandas as pd
from scipy.stats import kruskal
data = pd.read_csv('/content/drive/MyDrive/drive/awards.csv')
h_stat, p_val = kruskal('award_id','award_name','award_date')
print(f"kruskal-wallis Teat: H_stat={h_stat}, p_val={p_val}")
```
